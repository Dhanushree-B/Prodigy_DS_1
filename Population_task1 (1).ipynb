```python
import matplotlib.pyplot as plt
import numpy as np

# Sample age data for a population
ages = [22, 45, 36, 52, 23, 43, 33, 36, 26, 67, 78, 25, 29, 39, 40, 50, 60, 21, 24, 27, 30, 45, 49]

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=7, color='skyblue', edgecolor='black')
plt.title('Distribution of Age in Population')
plt.xlabel('Age Groups')
plt.ylabel('Number of People')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
